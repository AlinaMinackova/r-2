### Шаг 3. Формулировка ML-задачи на основе бизнес-задачи

добавим колонки с ценами за ночь и обслуживания с учетом сезона 

```python
def season(x):
    if x in ['January', 'February', 'December']:
        return 1
    if x in ['March', 'April', 'May']:
        return 1.2
    if x in ['June', 'July', 'August']:
        return 1.4
    return 1.2
```

```python
df['coef'] = df.arrival_date_month.apply(lambda x: season(x))
df
```

прибыль без депозита

```python
def profit(x):
    if(x['is_canceled'] == 0):
        if(x['reserved_room_type'] == 'A'):
            return (x['total_nights'] * 1000 * x['coef']) - ((x['total_nights'] // 2 + 1) * 400 * x['coef'])
        if(x['reserved_room_type'] == 'B'):
            return (x['total_nights'] * 800 * x['coef']) - ((x['total_nights'] // 2 + 1) * 350 * x['coef'])
        if(x['reserved_room_type'] == 'C'):
            return (x['total_nights'] * 600 * x['coef']) - ((x['total_nights'] // 2 + 1) * 350 * x['coef'])
        if(x['reserved_room_type'] == 'D'):
            return (x['total_nights'] * 550 * x['coef']) - ((x['total_nights'] // 2 + 1) * 150 * x['coef'])
        if(x['reserved_room_type'] == 'E'):
            return (x['total_nights'] * 500 * x['coef']) - ((x['total_nights'] // 2 + 1) * 150 * x['coef'])
        if(x['reserved_room_type'] == 'F'):
            return (x['total_nights'] * 450 * x['coef']) - ((x['total_nights'] // 2 + 1) * 150 * x['coef'])
        if(x['reserved_room_type'] == 'G'):
            return (x['total_nights'] * 350 * x['coef']) - ((x['total_nights'] // 2 + 1) * 150 * x['coef'])
    else:
        if(x['reserved_room_type'] == 'A'):
            return -((1000 + 400) * x['coef'])
        if(x['reserved_room_type'] == 'B'):
            return -((800 + 350) * x['coef'])
        if(x['reserved_room_type'] == 'C'):
            return -((600 + 350) * x['coef'])
        if(x['reserved_room_type'] == 'D'):
            return -((550 + 150) * x['coef'])
        if(x['reserved_room_type'] == 'E'):
            return -((500 + 150) * x['coef'])
        if(x['reserved_room_type'] == 'F'):
            return -((450 + 150) * x['coef'])
        if(x['reserved_room_type'] == 'G'):
            return -((350 + 150) * x['coef'])
```

```python
df['profit_one_order'] = df.apply(profit, axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(features)
features = pd.DataFrame(data=scaler.transform(features), columns=features.columns)

features
```

```python
# дерево решений
%matplotlib


from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

rez = []
best_rez = 0
best_model = None
dep = 0

for d in range(1, 20):
    model = DecisionTreeClassifier(random_state=12345, max_depth=d, class_weight='balanced')
    roc = cross_val_score(model, features, target, cv=8, scoring='roc_auc').mean()
    rez.append(roc)
    if(roc > best_rez):
        best_rez = roc
        best_model = model
        dep = d

print(best_rez, dep)
    
pd.DataFrame(rez, [i for i in range(1, 20)]).plot()
```
